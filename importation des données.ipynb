# Importation du jeux de donnée

# Trier pour ressortir le prix max et min

```python
import pandas as pd

# Aperçu des données Produits
products_data = pd.read_csv('products.csv')

# Trier le dataframe par la colonne 'price'
sorted_products = products_data.sort_values(by='price')

# Afficher les données triées
print(sorted_products)

# Trouver la valeur maximale dans la colonne 'price'
max_price = sorted_products['price'].max()
print("Max price:", max_price)

# Trouver la valeur minimale dans la colonne 'price'
min_price = sorted_products['price'].min()
display("Min price:", min_price)
```

# répartition par sex en %

```python
import pandas as pd

# Créer un DataFrame à partir des données fournies
data = customers_df

df = pd.DataFrame(data)

# Trier les clients par sexe
clients_par_sexe = df.groupby('sex')

# Compter le nombre de clients par sexe
nombre_clients_par_sexe = clients_par_sexe.size()

# Calculer le nombre total de clients
nombre_total_clients = nombre_clients_par_sexe.sum()

# Calculer le pourcentage de clients par sexe
pourcentage_par_sexe = (nombre_clients_par_sexe / nombre_total_clients) * 100

print("Pourcentage de clients par sexe :\n", pourcentage_par_sexe)
```

# Tri par session_id

```python
import pandas as pd

# Données de l'exemple
data =supp_invalide_date
df = pd.DataFrame(data)

# Trier le DataFrame par session_id
trie_invalide_date = df.sort_values(by='session_id')

display(trie_invalide_date)
```

# somme des prix par session_id

```python
import pandas as pd

# Données de l'exemple
data = trie_invalide_date
df = pd.DataFrame(data)

# Calculer la somme des prix par session_id
session_prices_sum = df.groupby('session_id')['price'].sum().reset_index()
session_prices_sum.columns = ['session_id', 'price_total_session']

# Fusionner le DataFrame d'origine avec le DataFrame des sommes de prix
sum_session_merged_df = pd.merge(df.drop(columns='price'), session_prices_sum, on='session_id').drop_duplicates(subset='session_id')

display(sum_session_merged_df)
```

# Tri par client_id

```python
import pandas as pd

# Données de l'exemple
data =sum_session_merged_df

df = pd.DataFrame(data)

# Trier le DataFrame par 'client_id'
tri_par_client_df = df.sort_values(by='client_id')

display(tri_par_client_df)
```

```python
import pandas as pd

# Données de l'exemple
data = tri_par_client_df
df = pd.DataFrame(data)

# Calculer la somme des prix par session_id
session_prices_sum = df.groupby('client_id')['price_total_session'].sum().reset_index()
session_prices_sum.columns = ['client_id', 'price_total_client']

# Fusionner le DataFrame d'origine avec le DataFrame des sommes de prix
total_df = pd.merge(df.drop(columns='price_total_session'), session_prices_sum, on='client_id').drop_duplicates(subset='client_id')

display(total_df)
```
